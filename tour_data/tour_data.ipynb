```python
class SalamBookingTours:

    def __init__(self):

        self.url_and_date_dict = url_and_date_object.url_to_date_dict


        self.cookies = {
            '_ga': 'GA1.1.194030406.1688460382',
            'COOKIES_CONSENT': 'CONSENT_ADS|CONSENT_ANALYSIS|CONSENT_TECHNICAL',
            '_ga_VK35BHX07G': 'GS1.1.1688460382.1.1.1688460425.0.0.0',
            '_ga_D1YQVGYNV7': 'GS1.1.1688460384.1.0.1688460425.19.0.0',
            'JSESSIONID': '7F93C3493D11A4F0FCB973ECE336BAA8.S005',
            'backend': 'production-tomee-server-1.travelc.internal:30005',
            'oam.Flash.RENDERMAP.TOKEN': '156x72jxvq',
        }

        self.headers = {
            'authority': 'salambooking.com',
            'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
            'accept-language': 'tr-TR,tr;q=0.9,en-US;q=0.8,en;q=0.7',
            'cache-control': 'no-cache',
            # 'cookie': '_ga=GA1.1.194030406.1688460382; COOKIES_CONSENT=CONSENT_ADS|CONSENT_ANALYSIS|CONSENT_TECHNICAL; _ga_VK35BHX07G=GS1.1.1688460382.1.1.1688460425.0.0.0; _ga_D1YQVGYNV7=GS1.1.1688460384.1.0.1688460425.19.0.0; JSESSIONID=7F93C3493D11A4F0FCB973ECE336BAA8.S005; backend=production-tomee-server-1.travelc.internal:30005; oam.Flash.RENDERMAP.TOKEN=156x72jxvq',
            'pragma': 'no-cache',
            'referer': 'https://salambooking.com/tr/2710/muhafazakar-turlar/moreideas',
            'sec-ch-ua': '"Not A(Brand";v="99", "Google Chrome";v="121", "Chromium";v="121"',
            'sec-ch-ua-mobile': '?0',
            'sec-ch-ua-platform': '"Windows"',
            'sec-fetch-dest': 'document',
            'sec-fetch-mode': 'navigate',
            'sec-fetch-site': 'same-origin',
            'sec-fetch-user': '?1',
            'upgrade-insecure-requests': '1',
            'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36',
        }


    def get_response(self, url):
        response = requests.get(url, headers=self.headers, cookies=self.cookies)
        return response
    
    def get_soup(self, response):
        soup = BeautifulSoup(response.text, "html.parser")
        return soup
    
    def get_inclusive(self, soup):
        inclusive = soup.find_all('span', class_='u-word-wrap--break closedtour-included-list dev-closedtour-included')
        inclusive_soup = BeautifulSoup(str(inclusive), "html.parser")
        inclusive = inclusive_soup.get_text(separator=".", strip=True)
        return inclusive
    
    def get_exclusive(self, soup):
        exclusive = soup.find_all('span', class_='u-word-wrap--break closedtour-included-list dev-closedtour-excluded')
        exclusive_soup = BeautifulSoup(str(exclusive), "html.parser")
        exclusive = exclusive_soup.get_text(separator=".", strip=True)
        return exclusive
    
    def get_title(self, soup):
        title = soup.find('title').text
        return title
    
    def get_tour_info(self, soup):
        tour_info = soup.find_all('div', class_='js-readmore-element')
        info_array = [info.text for info in tour_info]
        info_text = " ".join(info_array)
        return info_text
    
    def get_tour_price(self, soup):
        b = soup.find('b', class_='u-font-size--giant u-font-size--jumbo@sm clr--white o-block__item u-display--block u-line-height--1 dev-price-per-person')
        b = b.get_text( strip=True)
        return b
    
    def get_tour_dates(self, soup):
        tour_dates = soup.find('span', class_="c-tag c-tag--md dev-ribbon u-bdr-clr--transparent mt-10 ml-20 mr-20 u-font-size--default")
        return tour_dates
    
    # def start_process(self):
    #     df_list = []
    #     for url in self.tours_url_array:
    #         response = self.get_response(url)
    #         soup = self.get_soup(response)
    #         inclusive = self.get_inclusive(soup)
    #         exclusive = self.get_exclusive(soup)
    #         title = self.get_title(soup)
    #         tour_info = self.get_tour_info(soup)

    #         df = pd.DataFrame({
    #             "title": title,
    #             "inclusive": inclusive,
    #             "exclusive": exclusive,
    #             "tour_info": tour_info
    #         },index=[0])

    #         df_list.append(df)
    #     return pd.concat(df_list, ignore_index=True)

    def fetch_tour_data(self, url):
        response = self.get_response(url)
        soup = self.get_soup(response)
        inclusive = self.get_inclusive(soup)
        exclusive = self.get_exclusive(soup)
        title = self.get_title(soup)
        tour_info = self.get_tour_info(soup)
        price_info = self.get_tour_price(soup)
        tour_dates = self.url_and_date_dict[url]
        
        return {
            "title": [title],
            "tour_dates": [tour_dates],
            "inclusive": [inclusive],
            "exclusive": [exclusive],
            "price_info": [price_info],
            "tour_info": [tour_info]
        }

    def start_process(self):
        df_list = []
        # Use ThreadPoolExecutor to execute requests in parallel
        with concurrent.futures.ThreadPoolExecutor() as executor:
            # Schedule the fetch_tour_data function to be executed for each URL
            future_to_url = {executor.submit(self.fetch_tour_data, url): url for url in self.url_and_date_dict.keys()}
            for future in concurrent.futures.as_completed(future_to_url):
                try:
                    data = future.result()
                    df = pd.DataFrame(data)
                    df_list.append(df)
                except Exception as exc:
                    print(f'Generated an exception: {exc}')
        
        return pd.concat(df_list, ignore_index=True)
    
    def save_to_excel(self, df: pd.DataFrame, name: str):
        df.to_excel(f"{name}.xlsx", index=False)

            
```
