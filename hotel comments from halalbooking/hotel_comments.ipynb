```python
def get_name_country_date_room(soup):


    # Find the target element
    target_div = soup.find('div', class_='flag--container')

    # Extract and process the text
    # The texts are separated by commas and a <br> tag
    text_parts = target_div.get_text(separator='|').split('|')

    # Assign values to variables
    author_name = text_parts[0].strip()
    country = text_parts[2].strip()
    review_date = text_parts[4].strip()
    room_type = text_parts[5].strip()

    

    return author_name, country, review_date, room_type
```

```python
def get_review_text(soup: bs) -> list[str]:

    review_content = soup.find_all('div', class_='reviews-i--body')
    review_text = [review.get_text(separator='\n').strip() for review in review_content]
    return review_text
```

```python
from datetime import datetime

def convert_to_date(date_str):
    # Mapping of Turkish month abbreviations to month numbers
    months = {
        'Oca': 1, 'Şub': 2, 'Mar': 3, 'Nis': 4, 
        'May': 5, 'Haz': 6, 'Tem': 7, 'Ağu': 8, 
        'Eyl': 9, 'Eki': 10, 'Kas': 11, 'Ara': 12
    }

    # Split the string into month and year
    month_str, year = date_str.split()
    
    # Convert the month abbreviation to a number
    month = months.get(month_str)

    # Create a date object (assuming the first day of the month)
    if month:
        return datetime(year=int(year), month=month, day=1)
    else:
        raise ValueError("Invalid month abbreviation")
```

```python
def parse_comment_data(soup: bs) -> str:

    review_score = None
    review_header_translated = None
    review_header_original = None
    review_body_pros_translated = None
    review_body_pros_original = None
    review_body_cons_original = None
    review_body_cons_translated = None
    author_name = None
    country = None
    review_datetime = None
    room_type = None
    review_text = None

    # Extracting specific data
    review_score = soup.find('div', class_='reviews-i--score').text.strip()
    try:
 
        review_header_translated = soup.find('div', class_='reviews-i--header', attrs={'data-state': 'translated'}).text.strip()
        review_header_original = soup.find('div', class_='reviews-i--header', attrs={'data-state': 'original'}).text.strip()

        review_body_pros_translated = soup.find('div', class_='reviews-i--body reviews-i--body_pros js-dir-auto', attrs={'data-state': 'translated'}).text.strip()
        review_body_pros_original = soup.find('div', class_='reviews-i--body reviews-i--body_pros js-dir-auto', attrs={'data-state': 'original'}).text.strip()

        review_body_cons_translated = soup.find('div', class_='reviews-i--body reviews-i--body_cons js-dir-auto', attrs={'data-state': 'translated'}).text.strip()
        review_body_cons_original = soup.find('div', class_='reviews-i--body reviews-i--body_cons js-dir-auto', attrs={'data-state': 'original'}).text.strip()

        author_info = soup.find('div', class_='reviews-i--author').text.strip().split(',')
        author_name = author_info[0].strip()
        country = author_info[1].strip()
        
        review_footer = soup.find('div', class_='reviews-i--footer')

        # Extract the specific information
        review_date = review_footer.find_all('span', class_='comma')[-1].next_sibling.strip()
        review_datetime = convert_to_date(review_date)

        room_type = review_footer.find('div', class_='flag--container').contents[-1].strip()

    except:

        author_name, country, review_date, room_type = get_name_country_date_room(soup)
        review_text = get_review_text(soup)
        review_datetime = convert_to_date(review_date)

    current_date = datetime.now()
       
    one_year_ago = datetime(current_date.year - 1, current_date.month, current_date.day)



    if review_datetime < one_year_ago:
        print(f"review_datetime: {review_datetime} one_year_ago: {one_year_ago}")
        return None
        
    comment = {
        'author_name': author_name,
        'country': country,
        'review_datetime': review_datetime,
        'room_type': room_type,
        'review_text': review_text,
        'review_score': review_score,
        'review_header_translated': review_header_translated,
        'review_header_original': review_header_original,
        'review_body_pros_translated': review_body_pros_translated,
        'review_body_pros_original': review_body_pros_original,
        'review_body_cons_original': review_body_cons_original,
        'review_body_cons_translated': review_body_cons_translated,

    }

    return comment

    # Create a dictionary with the data
```

```python
def get_comments(soup):
    review_li = soup.find_all('li', class_=['reviews-i', 'reviews-i_expanded'])
    return review_li
```

```python
# Assuming `get_comments` and `process_comments` are defined functions

# Create a session object
#session = requests.Session()

comments_data_array = []
page = 0
resume = True

with tqdm(desc="Fetching Comments", unit="page") as pbar:
    while resume == True:
        try:
            #print(f"Fetching page {page}")
            params = {'page': str(page)}
            
            response = requests.get(url, headers=headers, params=params, cookies=cookies)
            soup = bs(response.content, 'html.parser')

            comments = get_comments(soup)

            if len(comments) == 0:
                print("No more comments")
                resume = False
            for comment in comments:
                try:
                    comment_data = parse_comment_data(comment)

                except:
                    print("Error parsing comment")
                    continue
      
                if comment_data == None:
                    print("No more comments")
                    resume = False
                    break
                print(comment_data)
                comments_data_array.append(comment_data)
            page += 1
            pbar.update(1)

        except requests.RequestException as e:
            print(f"Request failed: {e}")
            break

       
```
