```python
def fetch_data(check_in: str):

    check_out = check_in + timedelta(days=1)
    
    check_in = check_in.strftime('%Y-%m-%d %H:%M')
    check_out = check_out.strftime('%Y-%m-%d %H:%M')

    dates = f'"CHECKIN":"{check_in}","CHECKOUT":"{check_out}"'
    data = left_of_data + dates + rigt_of_data
    try:
        response = requests.post('https://4001.hoteladvisor.net/SP_PORTALV4_HOTELDETAILPRICE', params=params, headers=headers, data=data)
    except:
        print("encountered errow while fetching data")
        return None

    response = response.json()[0]
    df = pd.DataFrame(response)
    df['check_in_date'] = check_in
    df['check_out_date'] = check_out

    return df
```

```python
initial_date = datetime(year=2024, month=4, day=9, hour=0, minute=0)
last_date = datetime(year=2025, month=1, day=1, hour=0, minute=0)
```

```python
from datetime import datetime, timedelta
def datetime_range(start, end, delta=timedelta(days=1)):
    current = start
    while current < end:
        yield current
        current += delta

all_dates = list(datetime_range(initial_date, last_date))
```

```python
all_dates
```

```python
df_list = []

for check_in_date in tqdm(all_dates):

    response = fetch_data(check_in_date)

    df_list.append(response)





    
```

```python
df = pd.concat(df_list)
```
