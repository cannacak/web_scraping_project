```python
url_json = {
    "kadin_yeni_gelenler": {
        "headers": {
            "sec-ch-ua": "\"Chromium\";v=\"118\", \"Google Chrome\";v=\"118\", \"Not=A?Brand\";v=\"99\"",
            "Referer": "https://www.zara.com/tr/tr/kadin-yeni-gelenler-l1180.html?v1=2290613",
            "sec-ch-ua-mobile": "?0",
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36",
            "sec-ch-ua-platform": "\"Windows\""
        },
        "params": {
            "ajax": "true"
        },
        "url": "https://www.zara.com/tr/tr/category/2290613/products"
    },
    "kadin_dis_giyim": {
        "headers": {
            "sec-ch-ua": "\"Chromium\";v=\"118\", \"Google Chrome\";v=\"118\", \"Not=A?Brand\";v=\"99\"",
            "Referer": "https://www.zara.com/tr/tr/kadin-dish-giyim-l1184.html?v1=2290701",
            "sec-ch-ua-mobile": "?0",
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36",
            "sec-ch-ua-platform": "\"Windows\""
        },
        "params": {
            "ajax": "true"
        },
        "url": "https://www.zara.com/tr/tr/category/2290701/products"
    },
    "kadin_ceketler": {
        "headers": {
            "sec-ch-ua": "\"Chromium\";v=\"118\", \"Google Chrome\";v=\"118\", \"Not=A?Brand\";v=\"99\"",
            "Referer": "https://www.zara.com/tr/tr/kadin-ceketler-l1114.html?v1=2290770",
            "sec-ch-ua-mobile": "?0",
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36",
            "sec-ch-ua-platform": "\"Windows\""
        },
        "params": {
            "ajax": "true"
        },
        "url": "https://www.zara.com/tr/tr/category/2290770/products"
    },
    "kadin_blazerlar": {
        "headers": {
            "sec-ch-ua": "\"Chromium\";v=\"118\", \"Google Chrome\";v=\"118\", \"Not=A?Brand\";v=\"99\"",
            "Referer":  "https://www.zara.com/tr/tr/kadin-blazerlar-l1055.html?v1=2290737",
            "sec-ch-ua-mobile": "?0",
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36",
            "sec-ch-ua-platform": "\"Windows\""
        },
        "params": {
            "ajax": "true"
        },
        "url": "https://www.zara.com/tr/tr/category/2290737/products"
    },
    "kadin_yelekler": {
        "headers": {
            "sec-ch-ua": "\"Chromium\";v=\"118\", \"Google Chrome\";v=\"118\", \"Not=A?Brand\";v=\"99\"",
            "Referer":  "https://www.zara.com/tr/tr/kadin-dish-giyim-yelekler-l1204.html?v1=2290782",
            "sec-ch-ua-mobile": "?0",
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36",
            "sec-ch-ua-platform": "\"Windows\""
        },
        "params": {
            "ajax": "true"
        },
        "url": "https://www.zara.com/tr/tr/category/2290782/products"
    },
    "kadin_elbiseler": {
        "headers": {
            "sec-ch-ua": "\"Chromium\";v=\"118\", \"Google Chrome\";v=\"118\", \"Not=A?Brand\";v=\"99\"",
            "Referer":  "https://www.zara.com/tr/tr/kadin-elbiseler-l1066.html?v1=2290847",
            "sec-ch-ua-mobile": "?0",
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36",
            "sec-ch-ua-platform": "\"Windows\""
        },
        "params": {
            "ajax": "true"
        },
        "url": "https://www.zara.com/tr/tr/category/2290847/products"
    },
    "kadin_ust_giyim": {
        "headers": {
            "sec-ch-ua": "\"Chromium\";v=\"118\", \"Google Chrome\";v=\"118\", \"Not=A?Brand\";v=\"99\"",
            "Referer":  "https://www.zara.com/tr/tr/kadin-est-giyim-l1322.html?v1=2291012",
            "sec-ch-ua-mobile": "?0",
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36",
            "sec-ch-ua-platform": "\"Windows\""
        },
        "params": {
            "ajax": "true"
        },
        "url": "https://www.zara.com/tr/tr/category/2291012/products"
    },
    "kadin_triko": {
        "headers": {
            "sec-ch-ua": "\"Chromium\";v=\"118\", \"Google Chrome\";v=\"118\", \"Not=A?Brand\";v=\"99\"",
            "Referer":  "https://www.zara.com/tr/tr/kadin-erge-giyim-l1152.html?v1=2291084",
            "sec-ch-ua-mobile": "?0",
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36",
            "sec-ch-ua-platform": "\"Windows\""
        },
        "params": {
            "ajax": "true"
        },
        "url": "https://www.zara.com/tr/tr/category/2291084/products"
    },
    "kadin_gomlek_ve_bluz": {
        "headers": {
            "sec-ch-ua": "\"Chromium\";v=\"118\", \"Google Chrome\";v=\"118\", \"Not=A?Brand\";v=\"99\"",
            "Referer":  "https://www.zara.com/tr/tr/kadin-gemlekler-l1217.html?v1=2290933",
            "sec-ch-ua-mobile": "?0",
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36",
            "sec-ch-ua-platform": "\"Windows\""
        },
        "params": {
            "ajax": "true"
        },
        "url": "https://www.zara.com/tr/tr/category/2290933/products"
    },
    "kadin_t_shirt": {
        "headers": {
            "sec-ch-ua": "\"Chromium\";v=\"118\", \"Google Chrome\";v=\"118\", \"Not=A?Brand\";v=\"99\"",
            "Referer":  "https://www.zara.com/tr/tr/kadin-tishertler-l1362.html?v1=2290973",
            "sec-ch-ua-mobile": "?0",
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36",
            "sec-ch-ua-platform": "\"Windows\""
        },
        "params": {
            "ajax": "true"
        },
        "url": "https://www.zara.com/tr/tr/category/2290973/products"
    },
    "kadin_sweatshirt": {
        "headers": {
            "sec-ch-ua": "\"Chromium\";v=\"118\", \"Google Chrome\";v=\"118\", \"Not=A?Brand\";v=\"99\"",
            "Referer":  "https://www.zara.com/tr/tr/kadin-sweatshirtler-l1320.html?v1=2291105",
            "sec-ch-ua-mobile": "?0",
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36",
            "sec-ch-ua-platform": "\"Windows\""
        },
        "params": {
            "ajax": "true"
        },
        "url": "https://www.zara.com/tr/tr/category/2291105/products"
    },
    "kadin_pantolon": {
        "headers": {
            "sec-ch-ua": "\"Chromium\";v=\"118\", \"Google Chrome\";v=\"118\", \"Not=A?Brand\";v=\"99\"",
            "Referer":  "https://www.zara.com/tr/tr/kadin-pantolonlar-l1335.html?v1=2291147",
            "sec-ch-ua-mobile": "?0",
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36",
            "sec-ch-ua-platform": "\"Windows\""
        },
        "params": {
            "ajax": "true"
        },
        "url": "https://www.zara.com/tr/tr/category/2291147/products"
    },
    "kadin_kot_pantolon": {
        "headers": {
            "sec-ch-ua": "\"Chromium\";v=\"118\", \"Google Chrome\";v=\"118\", \"Not=A?Brand\";v=\"99\"",
            "Referer":  "https://www.zara.com/tr/tr/kadin-kot-pantolonlar-l1119.html?v1=2291201",
            "sec-ch-ua-mobile": "?0",
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36",
            "sec-ch-ua-platform": "\"Windows\""
        },
        "params": {
            "ajax": "true"
        },
        "url": "https://www.zara.com/tr/tr/category/2291201/products"
    },
    "kadin_etek": {
        "headers": {
            "sec-ch-ua": "\"Chromium\";v=\"118\", \"Google Chrome\";v=\"118\", \"Not=A?Brand\";v=\"99\"",
            "Referer":  "Referer': 'https://www.zara.com/tr/tr/kadin-etekler-l1299.html?v1=2291246",
            "sec-ch-ua-mobile": "?0",
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36",
            "sec-ch-ua-platform": "\"Windows\""
        },
        "params": {
            "ajax": "true"
        },
        "url": "https://www.zara.com/tr/tr/category/2291246/products"
    }
}


import requests
import pandas as pd
from concurrent.futures import ThreadPoolExecutor
from tqdm import tqdm


def fetch_product_data(response) -> pd.DataFrame:

    def extract_product_info(product):
        # Check if 'commercialComponents' is present and non-empty
        if 'commercialComponents' not in product or not product['commercialComponents']:
            print("commercialComponents key not found or is empty for this product.")
            return None

        # Extract product information
        product_info = product['commercialComponents'][0]
            # Skip products with "Kind" as "marketing"

        kind = product_info.get('kind', 'N/A').strip()
        name = product_info.get('name', 'N/A').strip()
        availability = product_info.get('availability', 'N/A').strip()
        
        if kind.lower() == 'marketing' or not name or not availability != 'N/A':
            return None

        price_value = product_info.get('price', 'N/A')
        price = f"{int(price_value) / 100} TL" if price_value != 'N/A' else 'N/A'

        # image_info = product_info['detail']['colors'][0]['xmedia'][0]
        # print(image_info)
        # image_url = f"https://static.zara.net/photos//{image_info['path']}/w/{image_info['id']}/{image_info['name']}.jpg?ts={image_info['timestamp']}"
        # print(image_url)
        # image_data = requests.get(image_url).content

        return {
            'Kind': product_info.get('kind', 'N/A'),
            'Name': product_info.get('name', 'N/A'),
            'Description': product_info.get('description', 'N/A'),
            'SectionName': product_info.get('sectionName', 'N/A'),
            'FamilyName': product_info.get('familyName', 'N/A'),
            'Price': price,
            'Availability': product_info.get('availability', 'N/A'),
            #'image': image_data  # You can download images if needed
               }

    try:
        # Make the HTTP request
        response.raise_for_status()
        result_json = response.json()

        # Extract product data from the response
        productGroups = result_json.get('productGroups', [])
        json_data = productGroups[0].get('elements', [])

        # Use ThreadPoolExecutor to parallelize the extraction process
        with ThreadPoolExecutor() as executor:
            product_list = list(filter(None, executor.map(extract_product_info, json_data)))

        return pd.DataFrame(product_list)

    except requests.RequestException as e:
        print(f"Error fetching data: {e}")
        return pd.DataFrame()

dataframes = []
for product in url_json:
    # Make the HTTP request 
    response = requests.get(url_json[product]['url'], params= url_json[product]['params'], headers=url_json[product]['headers'])

    # Fetch the product data
    df = fetch_product_data(response)

    df.insert(0,'Category', product)

    # Remove duplicates based on all columns
    df_no_duplicates = df.drop_duplicates()

    if not df_no_duplicates.empty:
        # Save the DataFrame as an HTML file
        dataframes.append(df_no_duplicates)
    break
all_data = pd.concat(dataframes)
```
