```python
def get_script_tag(soup):

    script_tag = soup.find('script', type='text/javascript', text=lambda t: t and 'window.insider_object.listing' in t)
    return script_tag
```

```python
import requests
from bs4 import BeautifulSoup
import json
from tqdm import tqdm

def extract_json_data(script_tag):
    try:
        script_content = script_tag.string
        json_data_start = script_content.index('{')
        json_data_end = script_content.rindex('}') + 1
        return json.loads(script_content[json_data_start:json_data_end])
    except (ValueError, json.JSONDecodeError) as e:
        print("Error parsing JSON data:", e)
        return None

adl_product_array = []

for url in tqdm(url_list):
    page = 1

    while True:
        print("Page:", page)
        response = requests.get(url, cookies=cookies, headers=headers, params={'page': page})
        soup = BeautifulSoup(response.text, 'html.parser')
        script_tag = get_script_tag(soup)

        if script_tag:
            product_data = extract_json_data(script_tag)
            if product_data and product_data.get("items"):
                for item in product_data["items"]:
               
                    product_attributes = {
                        "ID": item.get("id"),
                        "Name": item.get("name"),
                        "Price": item.get("unit_price"),
                        "Sale Price": item.get("unit_sale_price"),
                        "URL": item.get("url"),
                        "Stock": item.get("stock"),
                        "Color": item.get("color"),
                        "Size": item.get("size"),
                        "Image URL": item.get("product_image_url")
                    }
                    adl_product_array.append(product_attributes)
                    
            else:
                break
        else:
            break

        page += 1
```

```python
def get_product_attributes(url,session):
# Find the div with the product information
    
    response = session.get(url, cookies=cookies, headers=headers)
    
    soup = BeautifulSoup(response.text, 'html.parser')
    product_info_div = soup.find('div', id='collapsible-urun-bilgileri')
    
    # Extracting the desired information
    try:
        product_code = product_info_div.find('p').text.split(': ')[1]

    except:
        product_code = None
    # The attributes are directly under <p> tags or after <strong> tags within <p> tags
    attributes = []
    try:
        for p in product_info_div.find_all('p'):
            if p.strong:
                attribute_name = p.strong.text.strip().rstrip(':')
                attribute_value = p.text.replace(p.strong.text, '').strip()
                #print(attribute_name, attribute_value)
                attributes.append((attribute_name, attribute_value))
            else:
                attributes.append((p.text,))
    except:
        pass
            
    return product_code, attributes

```

```python
def get_product_details(attributes):
    attributes_dict = {}
    attribute_number = 0

    # Iterating through the attributes to populate the dictionary
    for attribute in attributes:
        if len(attribute) == 2:
            # When there is an explicit name and value for the attribute
            attribute_name, attribute_value = attribute
            attributes_dict[attribute_name] = attribute_value
        elif len(attribute) == 1:
            # When there is no explicit name, you might want to use a default name or handle differently
            # Here, we're just setting the attribute itself as a key with an empty value
            attributes_dict[f"attribute_{attribute_number}"] = attribute[0]
            attribute_number += 1

    return attributes_dict
```

```python
def create_session():
    # Create a session to keep track of the cookies
    session = requests.Session()
    session.cookies.update(cookies)
    session.headers.update(headers)
    return session
```

```python
from concurrent.futures import ThreadPoolExecutor
import concurrent
session = create_session()

def process_row(row):
    product_code, attributes = get_product_attributes(row.URL, session=session)
    product_details = get_product_details(attributes)
    product_details["product_code"] = product_code
    product_details["product_url"] = row.URL
    return product_details

product_details_array = []

# Wrapping your iteration with tqdm to display a progress bar
with ThreadPoolExecutor(max_workers=10) as executor:
    futures = {executor.submit(process_row, row): row for row in data.itertuples(index=False)}
    for future in tqdm(concurrent.futures.as_completed(futures), total=len(data)):
        product_details_array.append(future.result())
```
